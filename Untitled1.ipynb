```python
import pandas as pd
import plotly
import plotly.graph_objs as go
```

```python
data = pd.read_csv('data/popular_vote_by_state.csv', dtype={'dem_votes_2000':int,})
data = data.set_index('State')
static_dir = "slides/static/"
data.describe()
```

```python
state = 'WA'
go_dem = go.Bar(
    x=['2000', '2004', '2008', '2012', '2016'],
    y=data.loc[state][['dem_votes_2000', 'dem_votes_2004', 'dem_votes_2008', 'dem_votes_2012', 'dem_votes_2016']],
    name='Democrats',
    marker=dict(color='blue')
)
go_gop = go.Bar(
    x=['2000', '2004', '2008', '2012', '2016'],
    y=data.loc[state][['gop_votes_2000', 'gop_votes_2004', 'gop_votes_2008', 'gop_votes_2012', 'gop_votes_2016']],
    name='Republicans',
        marker=dict(color='red')

)
go_total = go.Scatter(
    x=['2000', '2004', '2008', '2012', '2016'],
    y=data.loc[state][['total_votes_2000', 'total_votes_2004', 'total_votes_2008', 'total_votes_2012', 'total_votes_2016']],
    yaxis='y2',
    name='Total Votes, Including 3rd party',
    
)
layout = go.Layout( 
    title=go.layout.Title(
        text=state+' Votes by Party',
    ),
    yaxis=dict(title='Party Votes'),
    yaxis2=dict(
        title='Total Votes, Including 3rd Party',
        titlefont=dict(
            color='green'
        ),
        tickfont=dict(
            color='green'
        ),
        overlaying='y1',
        side='right',
        range=[0, max(data.loc[state][['total_votes_2000', 'total_votes_2004', 'total_votes_2008', 'total_votes_2012', 'total_votes_2016']])*1.1]
    )
)
fig = go.Figure(data=[go_dem, go_gop, go_total], layout=layout)
plotly.offline.plot(fig, filename=static_dir+state+"/multiyear.html")
```
