```python
df_age = pd.read_csv('data/ACS_17_5YR_B01001A_with_ann.csv', encoding='iso-8859-1', skiprows=1)
df_age_compressed = pd.DataFrame()

df_age_compressed['Id2'] = df_age['Id2']
df_age_compressed['pop_estimate_2017'] = df_age['Estimate; Total:'] 
df_age_compressed['males_over_45'] = df_age['Estimate; Male: - 45 to 54 years'] + df_age['Estimate; Male: - 55 to 64 years'] + df_age['Estimate; Male: - 65 to 74 years']  + df_age['Estimate; Male: - 75 to 84 years']  + df_age['Estimate; Male: - 85 years and over']    
df_age_compressed['females_over_45'] = df_age['Estimate; Female: - 45 to 54 years'] + df_age['Estimate; Female: - 55 to 64 years'] + df_age['Estimate; Female: - 65 to 74 years'] + df_age['Estimate; Female: - 75 to 84 years'] + df_age['Estimate; Female: - 85 years and over']
df_age_compressed['males_over_55'] =  df_age['Estimate; Male: - 55 to 64 years'] + df_age['Estimate; Male: - 65 to 74 years']  + df_age['Estimate; Male: - 75 to 84 years']  + df_age['Estimate; Male: - 85 years and over']    
df_age_compressed['females_over_55'] =  df_age['Estimate; Female: - 55 to 64 years'] + df_age['Estimate; Female: - 65 to 74 years'] + df_age['Estimate; Female: - 75 to 84 years'] + df_age['Estimate; Female: - 85 years and over']
df_age_compressed['males_over_65'] = df_age['Estimate; Male: - 65 to 74 years']  + df_age['Estimate; Male: - 75 to 84 years']  + df_age['Estimate; Male: - 85 years and over']    
df_age_compressed['females_over_65'] = df_age['Estimate; Female: - 65 to 74 years'] + df_age['Estimate; Female: - 75 to 84 years'] + df_age['Estimate; Female: - 85 years and over']

df_age_compressed['males_over_45_rate'] = df_age_compressed['males_over_45'] / df_age_compressed['pop_estimate_2017']*100
df_age_compressed['females_over_45_rate'] = df_age_compressed['females_over_45'] / df_age_compressed['pop_estimate_2017']*100
df_age_compressed['males_over_55_rate'] = df_age_compressed['males_over_55'] / df_age_compressed['pop_estimate_2017']*100
df_age_compressed['females_over_55_rate'] = df_age_compressed['females_over_55'] / df_age_compressed['pop_estimate_2017']*100
df_age_compressed['males_over_65_rate'] = df_age_compressed['males_over_65'] / df_age_compressed['pop_estimate_2017']*100
df_age_compressed['females_over_65_rate'] = df_age_compressed['females_over_65'] / df_age_compressed['pop_estimate_2017']*100

df_age_compressed['people_over_45_rate'] = (df_age_compressed['males_over_45']+df_age_compressed['females_over_45'] ) / df_age_compressed['pop_estimate_2017']*100
df_age_compressed['people_over_55_rate'] = (df_age_compressed['males_over_55']+df_age_compressed['females_over_55'] ) / df_age_compressed['pop_estimate_2017']*100
df_age_compressed['people_over_65_rate'] = (df_age_compressed['males_over_65']+df_age_compressed['females_over_65'] ) / df_age_compressed['pop_estimate_2017']*100
df_age_compressed.drop(['males_over_45', 'females_over_45', 'males_over_55', 'females_over_55', 'males_over_65', 'females_over_65'],axis=1)
```

```python
# Pull in data
df = pd.read_csv('data/votes_2008_2016.csv')
df_education = pd.read_csv('data/Education.csv')[['FIPS Code', 'Percent of adults with less than a high school diploma, 2012-2016', 'Percent of adults with a high school diploma only, 2012-2016','2013 Urban Influence Code']]
df_unemployment = pd.read_csv('data/Unemployment.csv').dropna()
df_religion = pd.read_excel('data/U.S. Religion Census Religious Congregations and Membership Study, 2010 (County File) (2).xlsx')[['FIPS', 'TOTRATE', 'MPRTRATE', 'MPRTADH', 'POP2010', 'TOTADH', 'LDSRATE', 'EVANRATE', 'CATHRATE', 'ELCARATE']]
df_vets = pd.read_csv('data/vets_derived.csv')
# df_density = pd.read_csv('data/DEC_10_SF1_GCTPH1.US05PR_with_ann.csv',skiprows=1, encoding='iso-8859-1')
df['combined_fips'] = df.fips_code 
df_vets['vets_2016'] = df_vets['vets_2016'].str.replace(',','')
df_vets['vets_2016'] = df_vets['vets_2016'].astype(int)
# Clean up the number we intend to use
df_unemployment['Median_Household_Income_2016'] = df_unemployment['Median_Household_Income_2016'].str.replace('$', '') 
df_unemployment['Median_Household_Income_2016'] = df_unemployment['Median_Household_Income_2016'].str.replace(',', '')
df_unemployment['Median_Household_Income_2016'] = df_unemployment['Median_Household_Income_2016'].astype(int)

# Merge everything on FIPS
df = pd.merge(left=df, right=df_education, how='left', left_on='combined_fips', right_on='FIPS Code')
df = pd.merge(left=df, right=df_unemployment, how='left', left_on='combined_fips', right_on='FIPStxt')
df = pd.merge(left=df, right=df_religion, how='left', left_on='combined_fips', right_on='FIPS')
df = pd.merge(left=df, right=df_age_compressed, how='left', left_on='combined_fips', right_on='Id2')
df = pd.merge(left=df, right=df_vets, how='left', left_on='combined_fips', right_on='FIPS')

# df = pd.merge(left=df, right=df_density, how='left', left_on='combined_fips', right_on='Target Geo Id2')
# df = pd.merge(left=df, right=small_moves_dataset, how='left', left_on='combined_fips', right_on='FIPS')


# df = df[[
#     'per_gop_2016', 'per_dem_2016', 'per_gop_2012', 'per_dem_2012', 'Median_Household_Income_2016', 
#     'combined_fips', 'Percent of adults with less than a high school diploma, 2012-2016', 
#     'Percent of adults with a high school diploma only, 2012-2016', 'state_abbr',
#     'MPRTADH', 'POP2010', 'TOTADH'
# ]]
df = df.rename(index=str, columns={
    'total_2008': 'total_votes_2008',
    'total_2012': 'total_votes_2012',
    'total_2016': 'total_votes_2016',
    'gop_2016': 'votes_gop_2016',
    'gop_2012': 'votes_gop_2012',
    'gop_2008': 'votes_gop_2008',
    'dem_2016': 'votes_dem_2016',
    'dem_2012': 'votes_dem_2012',
    'dem_2008': 'votes_dem_2008',
    'oth_2016': 'votes_oth_2016',
    'oth_2012': 'votes_oth_2012',
    'oth_2008': 'votes_oth_2008',
})
df['per_dem_2016'] = df['votes_dem_2016']/df['total_votes_2016']
df['per_dem_2012'] = df['votes_dem_2012']/df['total_votes_2012']
df['per_dem_2008'] = df['votes_dem_2008']/df['total_votes_2008']
df['per_gop_2016'] = df['votes_gop_2016']/df['total_votes_2016']
df['per_gop_2012'] = df['votes_gop_2012']/df['total_votes_2012']
df['per_gop_2008'] = df['votes_gop_2008']/df['total_votes_2008']
df['state_abbr'] = df['State']
df['id'] = df['combined_fips']
```

```python
df_income = pd.read_csv('data/ACS_16_5YR_S1901_with_ann.csv', encoding='iso-8859-1', skiprows=1)[['Id2', 'Households; Estimate; Total', 'Families; Estimate; Total', 'Nonfamily households; Estimate; Total', 'Households; Estimate; Mean income (dollars)', 'Families; Estimate; Mean income (dollars)', 'Nonfamily households; Estimate; Mean income (dollars)']]
df_income = df_income.rename(
    index=str, 
    columns={
        'Households; Estimate; Total': '2016_households', 
        'Families; Estimate; Total': '2016_families', 
        'Nonfamily households; Estimate; Total': '2016_nonfamilies', 
        'Households; Estimate; Mean income (dollars)': '2016_household_income', 
        'Families; Estimate; Mean income (dollars)': '2016_family_income', 
        'Nonfamily households; Estimate; Mean income (dollars)': '2016_nonfamily_income'
    }
)
df_income['fips'] = df_income['Id2']
df_income['total_income'] = df_income['2016_households']*df_income['2016_household_income'] + df_income['2016_families']*df_income['2016_family_income'] + df_income['2016_nonfamilies']*df_income['2016_nonfamily_income']
df = pd.merge(left=df, right=df_income, how='left', left_on='combined_fips', right_on='fips')
```

```python
df['republican_shift'] = (df['per_gop_2016'] - df['per_gop_2012'])*100
df['rep_vote_change'] = (df['votes_gop_2016'] - df['votes_gop_2012'])/df['votes_gop_2012']*100
df['dem_vote_change'] = (df['votes_dem_2016'] - df['votes_dem_2012'])/df['votes_dem_2012']*100
df['protestant_pct'] = (df['MPRTADH']/df['POP2010'])*100
df['religious_pct'] = (df['TOTADH']/df['POP2010'])*100
df['positive_republican_shift'] = np.where(df['republican_shift']>0, df['republican_shift'], 0)
df['flipped_color'] = np.where(
    (df['per_gop_2016']>=df['per_dem_2016']) & (df['per_gop_2012']<=df['per_dem_2012']), 1, 0
)
df['High School Only or Less, 2012-2016'] = df['Percent of adults with less than a high school diploma, 2012-2016'] + df['Percent of adults with a high school diploma only, 2012-2016']
df['per_dem_2016_100%'] = df['per_dem_2016']*100
df['per_gop_2016_100%'] = df['per_gop_2016']*100
```
